# Splitting of Data for training and testing

```python
data.columns
```

```python
X=data[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement',
        'sqft_living15', 'sqft_lot15']]
```

```python
X
```

```python
Y=data['price']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.25,random_state=101)
```

# Normalization basically Standarization

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
```

```python
X_train_std=std.fit_transform(X_train)
X_test_std=std.transform(X_test)
```

# Model training

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train_std,Y_train)
```

```python
Y_pred=lr.predict(X_test_std)
```

```python
Y_pred
```

```python
Y_test
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
mean_absolute_error(Y_test,Y_pred)
```

```python
## Lets Check Using Other Regression Models
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(random_state=101)

# Fit the model to the training data
dt_model.fit(X_train_std, Y_train)

# Make predictions on the test data
dt_Y_pred = dt_model.predict(X_test_std)

# Calculate the mean absolute error and R-squared score
dt_mae = mean_absolute_error(Y_test, dt_Y_pred)
dt_r2 = r2_score(Y_test, dt_Y_pred)

print("Decision Tree Regression MAE:", dt_mae)
print("Decision Tree Regression R-squared:", dt_r2)
```

```python
  ## Random Forest Regression:
```

```python
from sklearn.ensemble import RandomForestRegressor

# Create a Random Forest Regression model
rf_model = RandomForestRegressor(random_state=101)

# Fit the model to the training data
rf_model.fit(X_train_std, Y_train)

# Make predictions on the test data
rf_Y_pred = rf_model.predict(X_test_std)

# Calculate the mean absolute error and R-squared score
rf_mae = mean_absolute_error(Y_test, rf_Y_pred)
rf_r2 = r2_score(Y_test, rf_Y_pred)

print("Random Forest Regression MAE:", rf_mae)
print("Random Forest Regression R-squared:", rf_r2)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# Create a Gradient Boosting Regression model
gb_model = GradientBoostingRegressor(random_state=101)

# Fit the model to the training data
gb_model.fit(X_train_std, Y_train)

# Make predictions on the test data
gb_Y_pred = gb_model.predict(X_test_std)

# Calculate the mean absolute error and R-squared score
gb_mae = mean_absolute_error(Y_test, gb_Y_pred)
gb_r2 = r2_score(Y_test, gb_Y_pred)

print("Gradient Boosting Regression MAE:", gb_mae)
print("Gradient Boosting Regression R-squared:", gb_r2)
```
